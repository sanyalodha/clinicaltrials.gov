## Extracting Data from API and converting it to a Pandas DataFrame

#### STUDIES EXTRACTED FOR THE YEAR 2015

```python
#Empty dictionary
data_2015 = {}

url = 'https://clinicaltrials.gov/api/query/study_fields'
# Looping through dates from 2015
for i in all_dates_2015:
    params_2015 = {'expr': ' AND '.join(["AREA[LocationCountry]United States", 
                                    f"AREA[StudyFirstPostDate]RANGE[{i}]", 
                                    "AREA[OverallStatus]Recruiting"]),
              'fields': ','.join(['NCTId','ConditionMeshTerm' , 'OverallStatus', 'Condition',
                                 'InterventionName', 'LocationCity', 
                                 'LocationState', 'LocationCountry', 
                                 'BriefSummary']),
              'max_rnk': 1000,
              'fmt': 'json'
             }
    # Converting dictionary item to url format
    params_2015 = urllib.parse.urlencode(params_2015)
    # GET data from API using base url and parameters in json format
    parsed_data_2015 = requests.get(url, params_2015).json()
    # Converting JSON TO FLAT TABLE
    base_2015_df = pd.json_normalize(parsed_data_2015['StudyFieldsResponse'], 
                           ['StudyFields'], 
                           ['APIVrs', 'DataVrs', 'Expression', 
                            'NStudiesAvail', 'NStudiesFound', 
                            'MinRank', 'MaxRank', 'NStudiesReturned'])
    data_2015.update({i: base_2015_df})

all_data_2015 = pd.concat(data_2015).droplevel(-1).reset_index()
all_data_2015
```

### CREATING FUNCTIONS


```python
# Function to pick the first term from a list of terms
def pick_first_term(x):
    if type(x) == list:
        return x[0]
    else:
        return x
#pick_first_term(['NCT02335671'])

# Function to replace the empty lists in our data
def replace_empty_list(x):
    if x ==[]:
        return 'N/A'
    else:
        return x
#replace_empty_list('N/A')   

# Function to convert strings into lowercase
def lowercase(x):
    return x.lower()
#lowercase('Breast Neoplasms')
```

## EDITING DATA

```python
# Retrieving only the studies that are recruiting

studies = data.assign(OverallStatus = data.get('OverallStatus').apply(pick_first_term))
studies = studies[studies.get('OverallStatus')=='Recruiting']

studies = studies.get(['NCTId', 'ConditionMeshTerm' ,'OverallStatus', 'Condition',
                                 'InterventionName', 'LocationCity', 
                                 'LocationState', 'LocationCountry', 
                                 'BriefSummary'])

# replacing the studies with an empty list as a mesh term with the term "N/A"
studies = studies.assign(ConditionMeshTerm = studies.get('ConditionMeshTerm').apply(replace_empty_list))

missing_items_list = studies[studies.get('ConditionMeshTerm')=='N/A'].index

studies
```

```python
# empty_studies: Dataframe with all the studies that do not have a mesh term

empty_studies = studies[studies.get('ConditionMeshTerm')=='N/A']

# Here, we pick the first term in the lists of CONDITIONS and we replace our empty mesh term slot with the term we've picked 
# from our field "CONDITION"
empty_studies = empty_studies.assign(ConditionMeshTerm = empty_studies.get('Condition').apply(pick_first_term))


empty_studies = empty_studies.assign(ConditionMeshTerm = empty_studies.get('ConditionMeshTerm').apply(lowercase), NCTId = empty_studies.get('NCTId').apply(pick_first_term))
empty_studies
```

## NOTE:

Now, the field "ConditionMeshTerm" contains lists of mesh terms for every study. So, in order to get one mesh term for each study, we apply the function "pick_first_term" to retrieve and store the first mesh term in the list as the only mesh term for the study in question.

```python

studies = studies.assign(ConditionMeshTerm =studies.get('ConditionMeshTerm').apply(pick_first_term),
                                           NCTId =studies.get('NCTId').apply(pick_first_term) )
studies = studies.assign(ConditionMeshTerm =studies.get('ConditionMeshTerm').apply(lowercase))

studies
```

```python
# Dropping studies with no mesh term so we can add the empty_studies dataframe later

incomplete_studies = studies.drop(missing_items_list)
incomplete_studies
```

```python
data_final = pd.concat([incomplete_studies,empty_studies])
data_final = data_final.sort_index()
data_final
```
